# Face Detection using Haar Cascades with OpenCV and Matplotlib

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('1000077167.jpg')  
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(image_rgb)
plt.title("Original Image")
plt.axis('on')
plt.show()
```

```python
roi = image[100:420, 200:550]  

mask = np.zeros_like(image)

mask[100:420, 200:550] = roi
```

```python
segmented_roi = cv2.bitwise_and(image, mask)
```

```python
segmented_roi_rgb = cv2.cvtColor(segmented_roi, cv2.COLOR_BGR2RGB)
plt.imshow(segmented_roi_rgb)
plt.title("Segmented ROI")
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python

image = cv2.imread('1000077167.jpg')  
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(image_rgb)
plt.title("Original Image")
plt.axis('off')
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
blurred_image = cv2.GaussianBlur(gray_image, (5, 5), 0)
```

```python
edges = cv2.Canny(blurred_image, 50, 150)
```

```python
plt.imshow(edges, cmap='gray')
plt.title("Canny Edge Detection")
plt.axis('off')
```

```python
contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
```

```python
result_image = image.copy() 
for contour in contours:
    if cv2.contourArea(contour) > 50:  
        x, y, w, h = cv2.boundingRect(contour)  
        cv2.rectangle(result_image, (x, y), (x + w, y + h), (0, 255, 0), 2)
```

```python
plt.imshow(cv2.cvtColor(result_image, cv2.COLOR_BGR2RGB))
plt.title("Handwriting Detection")
plt.axis('off')
```
